```python
#Scikit Learn
#supervised and unsupervised learning Algorithms
from sklearn.linear_model import LogisticRegression as lr
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
```

```python
#Extracting only independent variables
X=df.iloc[:, [8,10]].values# Credit history loan amount

#Extracting only the dependet Variables
y=df.iloc[:,12].values#loan Status
```

```python
## MACHINE LEARNING ##
# Splitting the dataset into training and testing dataset
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y, test_size=0.25, random_state=0)

#In this case we will be using the logistic Regression
#Logistic Regression is appropriate when the dependent variable is binary(0,1)
classifier= lr(random_state=0)
classifier.fit(X_train,y_train)
#Predicting the Test set Results
y_pred=classifier.predict(X_test)
y_pred

#Building the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm # Our model is 79.2% accurate, and 82.6% precise

#we can also find accuracy through a Python Module:
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
